```python
import pandas as pd
import plotly.express as px
```

```python
years = ["2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020", "2021", "2022", "2023"]
root = "C:/Users/uvm88/Desktop/Entregables/DATOS LIMPIOS 2013-2023/EGRESOS/CANCER/EGRESOSCANCER"
df_egresos_cancer = []

for year in years:
    df_base_cancer = pd.read_csv(root + year + ".csv")
    df_base_cancer.insert(0, "año", year)
    df_egresos_cancer.append(df_base_cancer)
df_concat = pd.concat(df_egresos_cancer)
df_concat = df_concat.reset_index().drop("index", axis = 1)
df_concat
```

```python
df_lineplot = df_concat["año"].value_counts().reset_index()
df_lineplot.columns = ["año", "cantidad"]
df_lineplot = df_lineplot.sort_values("año", ignore_index = True)
```

```python
fig_line = px.line(
              df_lineplot,
              x='año',
              y='cantidad',
              title='Linea temporal de la cantidad de egresos de cáncer (2013 - 2023)',
              template='plotly_white')


fig_line.update_layout(
    xaxis_title='Año',
    yaxis_title='Egresos',
    title_x=0.5 
)
fig_line.write_html("linea_temporal_egresos_cancer.html")
fig_line.show()
```

```python
df_concat = df_concat[(df_concat["EDAD"] < 112) & (df_concat["EDAD"] > 1)]

plotbox_m = px.box(
    df_concat,
    y = "EDAD",
    title = "Boxplot de egresos de cáncer por edad",
)

plotbox_m.update_layout(
    xaxis_title = "Personas",
    yaxis_title = "Edad",
    title_x = 0.5
)

plotbox_m.show()
plotbox_m.write_html("boxplot_egresos_cancer_por_edad.html")
```

```python
df_agrupado = df_concat.groupby("SEXO").size().reset_index(name="Cantidad")
mapeo = {
    1.0:"Hombres",
    2.0:"Mujeres",
}
df_agrupado["SEXO"] = df_agrupado["SEXO"].replace(mapeo)
df_agrupado = df_agrupado.drop([0, 3], axis = 0)
df_agrupado
```

```python
fig_defun = px.histogram(
    df_agrupado,
    x = "SEXO",
    y = "Cantidad",
    color = "SEXO",
    nbins = 2,
    barmode = "overlay",
    opacity = 0.6,
    title = "Distribución de egresos por género"
)

fig_defun.update_layout(
    xaxis_title="Género",
    yaxis_title="Egresos",
    title_x=0.5
)

fig_defun.show()
fig_defun.write_html("distribucion_egresos_cancer_genero.html")
```
