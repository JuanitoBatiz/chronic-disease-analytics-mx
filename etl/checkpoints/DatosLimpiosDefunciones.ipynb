```python
import pandas as pd
import os
import re

# Carpeta donde están los archivos limpios y renombrados (2013–2023)
ruta = "C:/Users/jesus/OneDrive/Desktop/Archivos de Salud 2010-2023/Datos Limpios/Datos Limpios Defunciones"

# Codigos CIE-10 a capturar:
#   - E10 hasta E14 (diabetes tipo 1–4)
#   - C00 hasta C99 (cáncer)
patron = re.compile(r'^(E1[0-4]|C[0-9]{2})', re.IGNORECASE)

for archivo in os.listdir(ruta):
    if not archivo.lower().endswith(".csv"):
        continue

    ruta_archivo = os.path.join(ruta, archivo)
    print(f" Filtrando por diabetes y cáncer en {archivo}…")

    df = pd.read_csv(ruta_archivo, encoding="utf-8")
    
    # Asegurarnos de que la columna existe y convertir a str
    if 'causa_defuncion' not in df.columns:
        print(f"    'causa_defuncion' no encontrada en {archivo}, omitiendo.")
        continue
    
    mask = df['causa_defuncion'].astype(str).str.match(patron)
    df_filtrado = df[mask].copy()

    # Guardar sobrescribiendo el archivo original
    df_filtrado.to_csv(ruta_archivo, index=False, encoding="utf-8")
    print(f"   {len(df_filtrado)} registros guardados en {archivo}")

print("  Todos los archivos contienen ahora solo muertes por diabetes o cáncer.")
```

```python
import pandas as pd
import os

# Ruta de tus archivos
ruta = r"C:\Users\jesus\OneDrive\Desktop\Archivos de Salud 2010-2023\Datos Limpios\Datos Limpios Defunciones"

# Años a procesar
anios = range(2013, 2024)

for anio in anios:
    nombre_archivo = f"DEFUN{anio}.csv"
    path_archivo = os.path.join(ruta, nombre_archivo)

    if not os.path.isfile(path_archivo):
        print(f" Archivo no encontrado: {nombre_archivo}")
        continue

    try:
        df = pd.read_csv(path_archivo)

        # Contar antes de modificar
        horas = ((df['edad'] >= 1000) & (df['edad'] < 2000)).sum()
        dias  = ((df['edad'] >= 2000) & (df['edad'] < 3000)).sum()
        meses = ((df['edad'] >= 3000) & (df['edad'] < 4000)).sum()
        total_modificados = horas + dias + meses

        # Conversión
        def convertir_selectivamente(valor):
            if 1000 <= valor < 2000:
                return valor / 8760
            elif 2000 <= valor < 3000:
                return valor / 365
            elif 3000 <= valor < 4000:
                return valor / 12
            else:
                return valor  

        df['edad'] = df['edad'].apply(convertir_selectivamente)
        df['edad_anios'] = df['edad']

        # Guardamos
        df.to_csv(path_archivo, index=False)

        print(f"{nombre_archivo} → {total_modificados} edades ajustadas ({horas} horas, {dias} días, {meses} meses)")

    except Exception as e:
        print(f" Error en {nombre_archivo}: {e}")
```

```python
import pandas as pd
import os

# Rutas
ruta_origen = r"C:\Users\jesus\OneDrive\Desktop\Archivos de Salud 2010-2023\Datos Limpios\Datos Limpios Defunciones"
ruta_destino = r"C:\Users\jesus\OneDrive\Desktop\DATOS LIMPIOS 2013-2023\DEFUNCIONES\DIABETESCANCER"

# Crear ruta final si no existe
os.makedirs(ruta_destino, exist_ok=True)

# codigos CIE‑10 a filtrar
prefijos_diabetes = [f"E1{d}" for d in range(0, 5)]              # E10–E14
prefijos_cancer   = [f"C{str(i).zfill(2)}" for i in range(0, 98)] # C00–C97

# Años a procesar
anios = range(2013, 2024)

for anio in anios:
    nombre_archivo = f"DEFUN{anio}.csv"
    path_archivo = os.path.join(ruta_origen, nombre_archivo)

    if not os.path.isfile(path_archivo):
        print(f" Archivo no encontrado: {nombre_archivo}")
        continue

    try:
        df = pd.read_csv(path_archivo)

        # Filtrar por los primeros 3 caracteres de la columna causa_defuncion
        df['causa_defuncion'] = df['causa_defuncion'].astype(str)

        df_diabetes = df[df['causa_defuncion'].str[:3].isin(prefijos_diabetes)]
        df_cancer   = df[df['causa_defuncion'].str[:3].isin(prefijos_cancer)]

        # Nombres de los archivos filtrados
        nombre_diabetes = f"DEFUNDIABETES{anio}.csv"
        nombre_cancer   = f"DEFUNCANCER{anio}.csv"

        # Guardar en la ruta destino
        df_diabetes.to_csv(os.path.join(ruta_destino, nombre_diabetes), index=False)
        df_cancer.to_csv(os.path.join(ruta_destino, nombre_cancer), index=False)

        print(f" {anio}: {len(df_diabetes)} diabetes | {len(df_cancer)} cáncer")

    except Exception as e:
        print(f" Error en {nombre_archivo}: {e}")
```
